### **Data Splitting**

```python
  train=walmart.head(147)
  test=walmart.tail(12)
```

```python
train
```

```python
test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# Linear Model
import statsmodels.formula.api as smf

linear_model=smf.ols('Footfalls~t',data=train).fit()# footfalls-target col,t- independent col
pred_linear=pd.Series(linear_model.predict(pd.DataFrame(test['t']))) # get Prediction for test data
rmse_linear=np.sqrt(mean_squared_error(test['Footfalls'],pred_linear))
rmse_linear # RMSE should be minimum
```

```python
# Exponential Model

exp_model=smf.ols('log_footfalls~t',data=train).fit()# footfalls-target col,t- independent col
pred_exp=pd.Series(exp_model.predict(pd.DataFrame(test['t']))) # get Prediction for test data
rmse_exp=np.sqrt(mean_squared_error(test['Footfalls'],pred_exp))
rmse_exp # RMSE should be minimum
```

```python
# Quadratic Model

quad_model=smf.ols('Footfalls~t+t_square',data=train).fit()# footfalls-target col,t- independent col
pred_quad=pd.Series(quad_model.predict(pd.DataFrame(test[['t','t_square']]))) # get Prediction for test data
rmse_quad=np.sqrt(mean_squared_error(test['Footfalls'],pred_quad))
rmse_quad # RMSE should be minimum
```

```python
# Additive Seasonality Model -pass from jan to Nov

add_sea=smf.ols('Footfalls~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=train).fit()# footfalls-target col,t- independent col
pred_add=pd.Series(add_sea.predict(pd.DataFrame(test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']]))) # get Prediction for test data
rmse_add=np.sqrt(mean_squared_error(test['Footfalls'],pred_add))
rmse_add # RMSE should be minimum
```

```python
# Additive Seasonalitywith quadratic Model -pass from jan to Nov

add_sea_quad=smf.ols('Footfalls~t+t_square+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=train).fit()# footfalls-target col,t- independent col
pred_addquad=pd.Series(add_sea_quad.predict(pd.DataFrame(test[['t','t_square','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']]))) # get Prediction for test data
rmse_addquad=np.sqrt(mean_squared_error(test['Footfalls'],pred_addquad))
rmse_addquad # RMSE should be minimum
```

```python
# multiplicative Seasonality Model -pass from jan to Nov

mul_sea=smf.ols('log_footfalls~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=train).fit()# footfalls-target col,t- independent col
pred_mul=pd.Series(mul_sea.predict(pd.DataFrame(test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']]))) # get Prediction for test data
rmse_mul=np.sqrt(mean_squared_error(test['Footfalls'],pred_mul))
rmse_mul # RMSE should be minimum
```

```python
# compare the results

data={"MODEL":pd.Series(['Linear','Exponential','Quadratic','Additive-sea','Additive-sea-quad','Multiplicative']),
      "RMSE_Values":pd.Series([rmse_linear,rmse_exp,rmse_quad,rmse_add,rmse_addquad,rmse_mul])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(["RMSE_Values"])

# RMSE value for additive seasonality with quadratic trend is giving less RMSE value so this model is good for this dataset
# To predict the footfall for next 12 months additive seasonality with quadratic trend model should be used
```
